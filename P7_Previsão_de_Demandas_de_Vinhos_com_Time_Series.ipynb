## Entendimento do Negócio

O vinho é uma bebida viva. Cada tipo de uva produz uma bebida com uma personalidade própria, que é ressaltada também pelo processo de produção. O resultado é uma bebida complexa, detalhada e cheia de nuances tanto no sabor quanto no aroma.

Por trazer tanta riqueza dentro da garrafa, carregando o trabalho e o cuidado de tantas pessoas, desde o plantio da uva até a venda, é preciso pensar em como armazenar  e transportar apropriadamente esse tipo de bebida. Nesse sentido, o lucro da empresa está diretamente associado à capacidade de logística no estoque, na guarda e no transporte, para que não se perca o vinho e o consumidor final receba um produto de qualidade.

<center><img alt="Vaga" width="50%" src="https://media.istockphoto.com/id/490390516/photo/pouring-red-wine.webp?b=1&s=170667a&w=0&k=20&c=Z0gxxmWXL0AgwmWGXoCgSV5WpvPUcg95XSVxJp4fzIo="></center>

Dentre os cuidados que se deve ter com o vinho destacam-se: a temperatura, níveis de umidade, exposição à luz, posicionamento e vibrações.

Vinhos são muito sensíveis a flutuações na temperatura. O calor excessivo acelera o processo de envelhecimento, enquanto o frio excessivo embota os sabores. É ideal manter a temperatura constante, entre 10 e 15 ºC, variando de acordo com o tipo de vinho.

A umidade deve ser mantida em torno de 70%, para evitar ressecamento da rolha, que causa oxidação e estraga a bebida. A exposição à luz UV degrada componentes químicos do vinho, piorando os sabores e os aromas. Por isso, é recomendado deixar as garrafas em locais escuros ou com baixa luminosidade.

Também é importante deixar as garrafas armazenadas horizontalmente, para manter a rolha sempre em contato com o vinho, garantindo a vedação adequada e evitando o ressecamento da rolha. Por fim, as vibrações (de máquinas e caminhões, por exemplo) devem ser evitadas para nao envelhecer o vinho antes da hora, pois perturba a deposição dos sedimentos e acelera as reações químicas.

Apesar de alguns vinhos ficarem melhores com o tempo, a maioria é de consumo "imediato" entre 1 e 3 anos. Com isso, fica claro que ter uma boa logística para armazenar e transportar os vinhos diferencia um negócio que vende daquele que empaca e perde o próprio estoque.

Portanto, fica claro que é preciso que a loja de vinho tenha todo o planejamento com o depósito, de modo a não ter prejuízos com perda de estoque e baixa satisfação dos consumidores.




## A Importância da Previsão de Demanda

A previsão de demanda diz respeito à capacidade de prever a demanda dos consumidores para um determinado produto. Trata-se de avaliar a disposição dos clientes de comprarem determinado produto a um preço específico.

Para isso, avaliamos os dados históricos de venda e, com base neles, projetamos as vendas e prevemos a demanda para determinado período no futuro.

A previsão de demanda está na base de uma operação funcional e lucrativa. Entre os pontos em que esse tipo de análise se destaca, estão:

- **Preparar o orçamento**

Este ponto trata da redução de riscos e de um melhor gerenciamento de recursos, através de decisões financeiras inteligentes. A previsão de demanda ajuda no planejamento para obter melhores margens de lucro, fluxo de caixa, alocação de recursos, timing para projetos de expansão, custos operacionais, de pessoal e gastos gerais.

- **Planejar e agendar a produção**

Caso a loja seja também a produtora do vinho, é relevante conhecer quando os clientes pedem por um produto para organizar a linha de produção deste produto. Determinados períodos podem ser marcados por demandas específicas, e isso favorece as vendas quando há estoque.

- **Planejamento de estoque**

O planejamento de estoque é crucial para um negócio de vinhos, pois trata de bebidas extremamente delicadas no seu manuseio e nos cuidados com armazenamento. O estoque deve ser planejado para que se priorize os produtos que possuem maior vazão de mercado, deixando uma menor proporção daqueles que não performam tão bem. Deve-se levar em conta que quanto maior for o estoque, mais caro é para mantê-lo. Assim, uma boa previsão de demanda ajuda a priorizar o que realmente gera valor.

- **Precificação e Promoções**

Aliado ao aperfeiçoamento do calendário de produção e do gerenciamento de estoque, a previsão de demanda ajuda também a capturar oportunidades de precificação e promoções. Junto com a previsibilidade vem o conhecimento de mercado e a identificação de oportunidades para crescer, seja formulando um esquema de preços competitivo, seja empregando estratégias de marketing direcionado.

Os preços dos vinhos cotados em reais não estão seguindo a cotação mais atualizada do dólar (dez/23). Para atualizar o conjunto de dados, vamos levar em conta a cotação mais recente de R\$ 4,88 para US$ 1,00.

Como algumas entradas na coluna `price_usd` utilizam a vírgula para separar os valores que contém milhar, será preciso retirá-la para, então, poder converter a coluna para um formato numérico.




```python
# Retirar a vírgula
df['price_usd'].replace(',', '', regex = True, inplace = True)

# Converter a coluna para float64
df['price_usd'] = df['price_usd'].astype('float64')

# Excluir a coluna dos valores em reais desatualizados
df.drop('price_brl', axis = 1, inplace = True)

# Criar uma nova coluna com valores default para os preços em reais
df['price_brl'] = 0

# Atualizando os valores, convertendo de dólar para reais segundo a cotação mais
# recente (dez/23)
df['price_brl'] = (df['price_usd'].values * 4.88).round(2)
```

## 3. Previsão de Demanda com Prophet


<center><img alt="Vaga" width="50%" src="https://cdn-images-1.medium.com/max/1250/1*Vbgsp_JEFAcJT7UhUD2KGQ.jpeg"></center>

A previsão de séries temporais é uma tarefa complexa que envolve a identificação de padrões em dados ordenados temporalmente. Para avaliar se esses padrões são de fato significativos e não apenas ruídos, é preciso realizar testes estatísticos sobre os dados. Para isso, existem diversos modelos que são configurados para testar as séries temporais, como o ARIMA (_Auto-Regressive Integrated Moving Average_), modelos auto-regressivos, suavização exponencial, além de técnicas baseadas em _Deep Learning_, como o LSTM (_long short term memory_).

Ao analisar os dados históricos, pode-se considerar uma única variável (_i.e_, as datas) para prever o _output_, caracterizando uma previsão univariada. Porém, se mais de um atributo é utilizado para realizar a previsão dos dados futuros, então a análise é dita multivariada.

De toda forma, os modelos de séries temporais irão buscar identificar dois indicadores essenciais:

- **Sazonalidade**: Repetições periódicas nos dados ao longo de um intervalo;
- **Tendência**: Direção preferencial dos dados ao longo do tempo (subida, descida ou estável).

O Prophet é um modelo cujos parâmetros são configurados para a tarefa de identificar esses dois pontos, fornecendo uma previsão de séries temporais. Para prosseguir com esse projeto, iremos utilizar o Prophet para previsão de demanda das lojas de vinho.

### 3.2 Testando a Estacionariedade

Nesta análise inicial, nota-se que os dados tem uma forte sazonalidade e uma tendência de crescimento gradual. Assim, a série temporal não parece ser estacionária, considerando que a média móvel tende a um aumento. A seguir, vamos realizar um teste estatístico para constatar a estacionariedade  ou não da série, através do teste ADF.

No teste ADF (_Augmented Dickey-Fuller_), a hipótese nula ($H_{0}$) é a de que a série não é estacionária e possui uma raíz única, ou seja, apresenta alguma estrutura de dependência temporal. Essa hipótese é rejeitada se conseguirmos constatar a estacionariedade da série com níveis de confiança maiores do que 90%, 95% ou 99%, correspondendo a p-valores de 0.1, 0.05 e 0.01, respectivamente.

A hipótese alternativa ($H_1$) constata a estacionariedade da série. Se o p-valor não tiver um nível de signficância menor do que 10%, então não haverá evidências para rejeitar a hipótese nula, de modo que os dados precisarão passar por alguns procedimentos (_e.g_, diferenciação) para tornar a série estacionária.

Vamos criar uma função para realizar o teste ADF em uma série temporal:

```python
def adf_test(time_series):

  print('Resultados do teste ADF\n')

  result = adfuller(time_series)

  df_output = pd.Series(
                        result[:2],
                        index = ['Estatística do Teste:', 'p-valor:']
                        )

  for key, value in result[4].items():
    df_output[f'Valor crítico {key}'] = value

  print(df_output)
```

```python
# Extraindo valores
X = df_ts['sales']
# Aplicando o teste ADF
adf_test(X)
```

A estatística ADF é o valor obtido no teste , quanto mais negativo for, maiores as chances de se rejeitar a $H_{0}$. Se olharmos os valores críticos, eles dizem respeito aos valores da estatística ADF que corresponderia aos níveis de confiança de 90%, 95% e 99%. Neste caso, obtivemos -2.36, enquanto que, para ultrapassar o limiar de 10% para o p-valor, precisaríamos de -2.56.

De fato, o p-valor é de 0.15, significando que não obtivemos evidências para rejeitar a hipótese nula de que nossa série é não-estacionária.

> **Resultado Teste ADF: não-estacionária**

Vamos aplicar um segundo teste estatístico para checar a estacionariedade. Trata-se do teste KPSS (_Kwiatkowski-Phillips-Schmidt-Shin_), que testa a estacionariedade baseada na tendência de uma série temporal.

É sempre bom aplicar dois testes estatísticos para saber se a série é realmente estacionária. Podemos ter alguns cenários:

- Os dois testes concordam na estacionariedade $\rightarrow$ A série é estacionária.
- Os dois testes concordam na não-estacionariedade $\rightarrow$ A série não é estacionária.
- O teste ADF indica estacionariedade e o KPSS não $\rightarrow$ A série é estacionária por diferenciação. Ou seja, a série temporal diferenciada deve ser testada por sua estacionariedade.
- O teste KPSS indica estacionariedade e o ADF não $\rightarrow$ A série é estacionária por tendência. A tendência precisa ser removida para tornar a série estritamente estacionária.

Vamos definir uma função para o teste KPSS:


```python
def kpss_test(time_series):

  print('Resultados do teste KPSS\n')

  result = kpss(time_series)

  df_output = pd.Series(
                        result[:2],
                        index = ['Estatística do Teste:', 'p-valor:']
                        )

  for key, value in result[3].items():
    df_output[f'Valor crítico {key}'] = value

  print(df_output)
```

```python
kpss_test(X)
```

Note que a estatística de teste deu um valor tão alto que está fora da tabela de p-valores. Assim, segundo o teste KPSS, estamos lidando com uma série estacionária com um nível de confiança que tende a 100%.

Com essa contradição entre os testes, temos a situação em que o KPSS indica estacionariedade e o ADF não. Sendo assim, precisamos trabalhar para retirar a tendência da série temporal para torná-la estritamente estacionária.

> **Resultado Teste KPSS: estacionária**

Outra maneira de se testar se a série é estacionária ou não é através do sumário estatístico, observando os parâmetros de média e variância em diferentes porções da série temporal. Vamos fazer isso a seguir, observando se em duas porções da série temporal há diferenças estatisticamente significativas na média e na variância.

```python
# Definindo o corte na metade da série
split = round(len(X) / 2)
# Associando variáveis para cada metade
X1, X2 = X[0: split], X[split:]
# Calculando as médias
mean_X1, mean_X2 = X1.mean(), X2.mean()
# Calculando as variâncias
var_X1, var_X2 = X1.var(), X2.var()

print('Sumário Estatístico:\n')

print(f'Média',
      f'\nX1: {round(mean_X1, 2)}',
      f'\nX2: {round(mean_X2, 2)}',
      f'\nVariação: {round((mean_X2 - mean_X1) * 100/ mean_X1, 2)}%')

print(f'\nVariância',
      f'\nX1: {round(var_X1, 2)}',
      f'\nX2: {round(var_X2, 2)}',
      f'\nVariação: {round((var_X2 - var_X1) * 100/ var_X1, 2)}%')
```

Note que, enquanto a média varia ligeiramente, em torno de 2,5%, a variância muda drasticamente, em torno de 24,1%. Se considerássemos apenas a média, essa flutuação no valor seria aceitável para considerar a série como estacionária. Porém, devido à variância, vemos que ela realmente não pode ser estacionária.

_PS: Importante notar que esse método de tirar a média de diferentes pedaços da série temporal não é muito seguro. Para definir a estacionariedade, o melhor a se fazer são testes estatísticos mais robustos, como o ADF e KPSS._

> **Conclusão: Série Não-Estacionária por Tendência**

### 3.6 Convertendo em Série Estacionária

Como vimos na Seção 3.2, o teste ADF indicou que a série temporal não é estacionária, enquanto que o KPSS indicou a estacionariedade. Devido a isso, foi possível concluir que é necessário **remover a tendência** da série de modo que ela se torne estritamente estacionária e os dois testes concordem.

```python
# Utilizando o df da série não-estacionária com datas como índice
df_not_st.head()
```

Para calcular a média móvel simples da série temporal, iremos utilizar o método `rolling()` do Pandas, onde colocamos como argumento o tamanho da intervalo que queremos medir. Em seguida, tomamos a média.

```python
fig, ax = plt.subplots(figsize = (8, 4))

df_not_st.plot(color = 'darkblue', ax = ax)
# Média móvel de 30 dias
df_not_st.rolling(30).mean().plot(ax = ax, linewidth = 2)

ax.legend(['Dados observados', 'Média móvel (30 dias)'])
ax.set_title('Média móvel mensal')
ax.set_yticks(np.arange(38e3, 52e3, 4e3))
ax.set_xlabel('')
ax.set_ylabel('Vendas')

ax.set_xticks(
              ticks = ax.get_xticks().tolist(),
              labels = labels_1,
              rotation = 45
              )

fig.tight_layout()
fig.show()
```

Para remover a tendência e diminuir os efeitos da sazonalidade sobre a série temporal, vamos aplicar o logaritmo aos dados de vendas e à média móvel. Em seguida, iremos subtrair essas duas componentes calculadas, que se encontram na mesma escala, de modo que teremos como resultado uma série aproximadamente estacionária.

```python
# Aplicando o logaritmo sobre a série temporal e média móvel
df_log = np.log(df_not_st)
ma_log = np.log(df_not_st.rolling(30).mean())

# Construindo um novo conjunto de dados baseado na diferença dos dados logarítmicos
df_detrend = (df_log - ma_log).dropna()
# Média móvel do conjunto normalizado
ma_detrend = df_detrend.rolling(30).mean()
# Desvio-padrão móvel do conjunto normalizado
std_detrend = df_detrend.rolling(30).std()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))

df_detrend.plot(color = 'darkblue', ax = ax)
ma_detrend.plot(linewidth = 2, ax = ax)
std_detrend.plot(color = 'yellow', linewidth = 2, ax = ax)

ax.set_title('Série Estacionária')

ax.legend(['Dados observados',
           'Média móvel (30 dias)',
           'Desvio-padrão móvel (30 dias)'])

ax.set_xlabel('')
ax.set_ylabel('')

ax.set_xticks(
              ticks = ax.get_xticks().tolist(),
              labels = labels_1,
              rotation = 45
              )

fig.tight_layout()
fig.show()
```

Visualmente, a série já parece ser estacionária. Se compararmos com os gráficos obtidos ao final da Seção 3.1, vemos que os padrões dos dados normalizados se parecem muito com os padrões residuais.

Note que agora a média móvel oscila, mas num intervalo muito menor. Além disso, o desvio-padrão (raiz quadrada da variância), também não varia muito.

Vamos realizar o teste ADF para confirmar a estacionariedade:

```python
# Extraindo valores
X_detrend = df_detrend['y'].values
# Realizando o teste
adf_test(X_detrend)
```

Com esse teste, obtivemos uma estatística de -4,82 (comparado com -2,36 do primeiro). O p-valor, que representa nosso nível de significância para aceitar a hipótese alternativa, está muito abaixo do limiar de 1%, o que significa que podemos rejeitar a hipótese nula com quase 100% de confiança. Assim, a série é estacionária pelo teste ADF.

Vamos ver os resultados para o teste KPSS:

```python
kpss_test(X_detrend)
```

O teste KPSS também concorda que a série é estacionária, retornando uma estatística de 0,54 que corresponde a um nível de confiança de 97%.

Como os dois testes concordam entre si, a série temporal agora é estacionária.

- **Testes de Estacionariedade**

Num primeiro momento, testamos a estacionariedade da série temporal através dos testes ADF e KPSS. Os testes apresentam abordagens distintas para testar se a série é estacionária ou não, e segundo os resultados de cada um, podemos diagnosticar se há estacionariedade ou não. No nosso caso, houve uma discordância entre os testes, de modo que constatamos que a série é não-estacionária por tendência.

- **Ajuste sobre a Série Não-estacionária**

Ainda assim, fizemos um primeiro ajuste do modelo sobre a série não-estacionária, de modo a termos um _baseline_ para comparações. Os dados históricos foram muito bem ajustados pelo modelo, porém o erro absoluto médio (MAE = 1015,2) foi muito alto. Para termos boas previsões futuras, era preciso diminuir esse erro através do ajuste na série estacionária.

- **Convertendo a Série Temporal**

Como a série é não-estacionária por tendência, segundo nossos testes estatísticos, é preciso aplicar uma transformação nos dados para que eles fiquem numa escala onde as variações dos parâmetros estatísticos se tornem desprezíveis. Para isso, aplicamos o logarítmo aos dados e sobre a média móvel, de modo que obtivemos uma série estacionária, com ambos os testes ADF e KPSS concordando na estacionariedade.

- **Ajuste do Modelo sobre a Série Estacionária**

Com o modelo ajustado sobre os dados estacionários, obtivemos o menor erro absoluto médio (MAE = 0,74) e o menor erro médio absoluto percentual (MAPE = 0,07), o que confirmou a acurácia do modelo.

- **Previsões Futuras e Limites do Modelo**

Realizando a previsão sobre os dados futuros, conseguimos prever as vendas com um nível de confiança alto, e uma incerteza que se manteve dentro da oscilação esperada das vendas diárias.

Ao testar os limites do modelo, prevendo dois anos de vendas no futuro, começamos a ver os limites do intervalo de incerteza aumentar rapidamente, como esperado.

**Em suma**: O modelo apresentou uma capacidade preditiva alta da demanda de vendas de vinhos até um ano no futuro. Com um erro absoluto médio (**MAE**) baixíssimo, de **0,74**, e um **MAPE abaixo de 10%**, trata-se de um modelo robusto, capaz de generalizar para novos dados no futuro e prever a demanda com precisão.
